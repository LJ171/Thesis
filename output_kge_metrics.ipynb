```python
import os
```

```python
datasets = ['codex-m', 'fb15k-237', 'wnrr', 'yago3-10']
models = [('complex', 'ComplEx'), ('conve', 'ConvE'), ('rescal', 'RESCAL')]
```

```python
lines = ['hits_at_1_filtered:', 'hits_at_10_filtered:', 'mean_reciprocal_rank_filtered:']

for dataset in datasets:
    for model, model_name in models:
        print()
        print()
        print()
        print(f'{dataset} - {model_name}')
        for i in range(1,6):
            if os.path.exists(os.path.join('experiments', f'{dataset}_{model}_{i}')):
                with open(os.path.join('experiments', f'{dataset}_{model}_{i}', 'test_results','kge.log')) as f:
                    count=0
                    for j, line in enumerate(f):
                        if any(l in line for l in lines): 
                            print(line[37:-1])
                            count+=1
                            if(count>=len(lines)):
                               break
            print()
```
